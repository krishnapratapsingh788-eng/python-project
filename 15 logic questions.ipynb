```python

# 1. Prime Number Checker
def is_prime(n):
    if n <= 1:
        return "Not Prime"
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return "Not Prime"
    return "Prime"
print(is_prime(22))
print(is_prime(17))
```

```python
# 2. Armstrong Number
def is_armstrong(n):
    total = sum(int(digit)**3 for digit in str(n))
    return "Armstrong" if total == n else "Not Armstrong"
print(is_armstrong(153))
print(is_armstrong(27))
```

```python
# 3. Palindrome Number or String
def is_palindrome(value):
    s = str(value)
    return "Palindrome" if s == s[::-1] else "Not Palindrome"
print(is_palindrome("mam"))
print(is_palindrome("harshit"))
```

```python
#4. Factorial Calculator
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Example
print(factorial(5))  
```

```python
#5. Fibonacci Series
def fibonacci(n):
    a, b = 0, 1
    series = []
    for _ in range(n):
        series.append(a)
        a, b = b, a + b
    return series

# Example
print(fibonacci(6))
```

```python
#6. Sum Of Natural Numbers
def sum_natural(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

# Example
print(sum_natural(10)) 
```

```python
#7.Reverse a Number
def reverse_number(n):
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return rev

# Example
print(reverse_number(1234)) 
```

```python
#8. GCD (Greatest Commmon Divisor)
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Example
print(gcd(12, 18))  
```

```python
#9. LCM (Least Common Multiple)
def lcm(a, b):
    def gcd(x, y):
        while y:
            x, y = y, x % y
        return x
    return abs(a * b) // gcd(a, b)

# Example
print(lcm(4, 6))  
```

```python
#10. Sum of Digits 
def sum_digits(n):
    if n == 0:
        return 0
    return n % 10 + sum_digits(n // 10)

# Example
print(sum_digits(456))  
```

```python
#11. Number Reversal Palindrome
def is_number_palindrome(n):
    original = n
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return "Palindrome" if original == rev else "Not Palindrome"

# Example
print(is_number_palindrome(121))  
```

```python
#12. Count Factors of a Number
def count_factors(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    return count

# Example
print(count_factors(12))  
```

```python
#13. Perfect Number
def is_perfect(n):
    total = sum(i for i in range(1, n) if n % i == 0)
    return "Perfect Number" if total == n else "Not Perfect"

# Example
print(is_perfect(28)) 
```

```python
#14. swap two number without temp
def swap(a, b):
    a = a + b
    b = a - b
    a = a - b
    return a, b

# Example
a, b = swap(5, 10)
print("a =", a, "b =", b)  
```

```python
#15. count digit in a number
def count_digits(n):
    count = 0
    while n > 0:
        count += 1
        n //= 10
    return count

# Example
print(count_digits(12345))
```
