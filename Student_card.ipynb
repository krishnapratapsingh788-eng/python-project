```python
#Create a small program using class & objects to store a studentâ€™s marks and calculate:
#.Total marks
#.Average marks
#.Grade
```

```python
class Student:
    # Constructor to store name and marks
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks  # list of marks

    # Method to calculate total
    def total_marks(self):
        return sum(self.marks)

    # Method to calculate average
    def average_marks(self):
        return self.total_marks() / len(self.marks)

    # Method to find grade
    def grade(self):
        avg = self.average_marks()
        if avg >= 90:
            return "A+"
        elif avg >= 80:
            return "A"
        elif avg >= 70:
            return "B"
        elif avg >= 60:
            return "C"
        else:
            return "D"

# ----- Main Program -----

# Creating an object
student1 = Student("Krishna", [85, 90, 78, 92, 88])

# Printing Report
print("----- Student Report Card -----")
print("Name:", student1.name)
print("Total Marks:", student1.total_marks())
print("Average Marks:", student1.average_marks())
print("Grade:", student1.grade())
```
