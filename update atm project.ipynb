```python
#### ATM Machine Simulator
print("Welcome to the ATM!")
balance = 0

# User data
user = {
    "101": {
        "name": ["krishna", "Anant", "Akshat"],
        "pin": ["8736", "2146", "2536"]
    }
}

# Login system
while True:
    card_no = input("Enter your card number: ")
    if card_no in user:
        pin = input("Enter your PIN: ")
        if pin in user[card_no]["pin"]:
            print("PIN verified successfully!")
            name = input("Enter your name: ")
            if name in user[card_no]["name"]:
                print("Name verified successfully")
                break   # Exit login loop
            else:
                print("Name does not match. Please try again.")
        else:
            print("Invalid PIN. Please try again.")
    else:
        print("Invalid card number. Please try again.")

# ATM functions
def deposit(amount):
    global balance
    if amount > 0:
        balance += amount
        print(f"Deposit successful! Current Balance = {balance}")
    else:
        print("Invalid deposit amount. Please enter a positive value.")

def withdraw(amount):
    global balance
    if amount > balance:
        print(f"Insufficient Balance! Your current balance is {balance}")
    elif amount <= 0:
        print("Invalid withdrawal amount.")
    else:
        balance -= amount
        print(f"Withdrawal successful! Current Balance = {balance}")

def check_balance():
    print(f"Your current balance is {balance}")

def atm():
    while True:
        print("\nMenu:")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            amt = int(input("Enter amount to deposit: "))
            deposit(amt)
        elif choice == "2":
            amt = int(input("Enter amount to withdraw: "))
            withdraw(amt)
        elif choice == "3":
            check_balance()
        elif choice == "4":
            print("Thank you for using our ATM. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

atm()
```
